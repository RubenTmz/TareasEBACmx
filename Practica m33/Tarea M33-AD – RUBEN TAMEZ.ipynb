VENTAS

```python
# 1) Total de ventas por año
df.groupby('Year')['Global_Sales'].sum()
```

```python
# 2) Juegos más vendidos por año
df.groupby(['Year','Name'])['Global_Sales'].sum().sort_values(ascending=False)
```

```python
#3) Plataformas con más ingresos por año. Todos estos números desglosados por ingreso: NA, EU, JP, Otros y el Global
df.groupby(['Year','Platform'])[['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']].sum().sort_values(by='Global_Sales', ascending=False)
```

```python
#4) Gráfico de comparacion de ventas entre géneros por año
df.groupby(['Year','Genre'])['Global_Sales'].sum().unstack().plot(kind='bar')
```

```python
# Insights: 2 Gráficos adicionales interesantes

# 1) Gráfico de ventas globales por plataforma a lo largo del tiempo
df_platform_sales = df.groupby(['Year', 'Platform'])['Global_Sales'].sum().unstack()
df_platform_sales.plot(kind='line', figsize=(12, 6))
plt.title('Ventas Globales por Plataforma a lo largo del Tiempo')
plt.xlabel('Año')
plt.ylabel('Ventas Globales (millones)')
plt.legend(title='Plataforma', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 2) Gráfico de los 10 principales editores por ventas globales
df_publisher_sales = df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)
df_publisher_sales.plot(kind='bar', figsize=(12, 6), color='skyblue')
plt.title('Top 10 Marcas por Ventas Globales')
plt.xlabel('Editor')
plt.ylabel('Ventas Globales (millones)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Análisis Exploratorio de Datos (EDA)

```python
# 1) Qué género ha tenido el mayor y menor crecimiento

# Filtrar últimos 4 años
df_ultimos4 = df[df['Year'] >= df['Year'].max() - 3]

ventas_genero = df_ultimos4.groupby('Genre')['Global_Sales'].sum().sort_values()

ventas_genero.plot(kind='barh', color='lightgreen', figsize=(10,5))
plt.title('Ventas por Género (últimos 4 años)')
plt.xlabel('Ventas Globales (millones)')
plt.tight_layout()
plt.show()
```

```python
# 2) Qué plataforma ha tenido el mayor y menor crecimiento

ventas_plataforma = df_ultimos4.groupby('Platform')['Global_Sales'].sum().sort_values()

ventas_plataforma.head(10).plot(kind='barh', figsize=(10,5), color='lightblue')
plt.title('Plataformas con Menores Ventas (últimos 4 años)')
plt.xlabel('Ventas Globales (millones)')
plt.tight_layout()
plt.show()
```

```python
# 3) Qué videojuego ha tenido el mayor y menor crecimiento

ventas_juegos = df_ultimos4.groupby('Name')['Global_Sales'].sum().sort_values()

ventas_juegos.head(10).plot(kind='barh', figsize=(10,5), color='salmon')
plt.title('Juegos con Menores Ventas (últimos 4 años)')
plt.xlabel('Ventas Globales (millones)')
plt.tight_layout()
plt.show()
```

```python
# 2 elementos adicionales que muestren oportunidades de mercado.
# 1️° Oportunidad: Géneros con ventas crecientes, pero baja participación actual
genero_total = df_ultimos4.groupby('Genre')['Global_Sales'].sum().sort_values()
genero_total.plot(kind='barh', figsize=(10,5), color='lightgreen')
plt.title('Ventas globales por género (últimos 4 años)')
plt.xlabel('Ventas (millones)')
plt.tight_layout()
plt.show()
# Insight: Géneros como Puzzle o Adventure podrían tener espacio para crecimiento si se impulsan

# 2️° Oportunidad: Plataformas con repunte reciente en ventas
ventas_anio_plataforma = df_ultimos4.groupby(['Year', 'Platform'])['Global_Sales'].sum().unstack(fill_value=0)
ventas_anio_plataforma.T.plot(kind='bar', stacked=True, figsize=(12,6))
plt.title('Ventas por Plataforma en los Últimos 4 Años')
plt.ylabel('Ventas Globales (millones)')
plt.xlabel('Plataforma')
plt.tight_layout()
plt.show()
# Insight: Algunas plataformas antiguas o secundarias muestran repuntes que podrían representar nichos rentables
```
