# 6.- Reportes por grupos : estado civil, educación, trabajo – cuántos registros

```python
print("=" * 60)
print("REPORTES POR GRUPOS")
print("=" * 60)

# Función para generar reportes detallados por categoría
def generar_reporte_por_grupo(df, columna, nombre_columna):
    print(f"\n{'='*50}")
    print(f"ANÁLISIS POR {nombre_columna.upper()}")
    print(f"{'='*50}")
    
    # Conteo de registros por categoría
    conteo = df[columna].value_counts()
    porcentaje = df[columna].value_counts(normalize=True) * 100
    
    # Crear dataframe con el reporte
    reporte = pd.DataFrame({
        'Categoría': conteo.index,
        'Cantidad': conteo.values,
        'Porcentaje': porcentaje.values.round(2)
    })
    
    # Mostrar tabla
    print(f"\nDistribución de registros por {nombre_columna}:")
    print(reporte.to_string(index=False))
    
    # Estadísticas adicionales
    print(f"\nEstadísticas adicionales:")
    print(f"Total de registros: {len(df)}")
    print(f"Número de categorías únicas: {df[columna].nunique()}")
    print(f"Valor más frecuente: {conteo.index[0]} ({conteo.values[0]} registros)")
    
    # Distribución de la variable objetivo 'y' por categoría
    if 'y' in df.columns:
        print(f"\nDistribución de la variable objetivo 'y' por {nombre_columna}:")
        cross_tab = pd.crosstab(df[columna], df['y'], normalize='index') * 100
        cross_tab.columns = ['No' if col == 'no' else 'Sí' for col in cross_tab.columns]
        cross_tab['Total'] = df[columna].value_counts()
        print(cross_tab.round(2))
    
    return reporte

# Generar reportes para cada variable categórica
reporte_marital = generar_reporte_por_grupo(df, 'marital', 'estado civil')
reporte_education = generar_reporte_por_grupo(df, 'education', 'educación')
reporte_job = generar_reporte_por_grupo(df, 'job', 'trabajo')

# Reporte adicional: distribución por estado civil y educación
print(f"\n{'='*60}")
print("DISTRIBUCIÓN CRUZADA: ESTADO CIVIL vs EDUCACIÓN")
print(f"{'='*60}")

cross_marital_education = pd.crosstab(df['marital'], df['education'])
print(cross_marital_education)

# Visualizaciones
print(f"\n{'='*60}")
print("VISUALIZACIONES")
print(f"{'='*60}")

# Configurar el estilo de las gráficas
plt.style.use('seaborn-v0_8')
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Distribución de Variables Categóricas', fontsize=16, fontweight='bold')

# Gráfica 1: Estado civil
conteo_marital = df['marital'].value_counts()
axes[0, 0].pie(conteo_marital.values, labels=conteo_marital.index, autopct='%1.1f%%', startangle=90)
axes[0, 0].set_title('Distribución por Estado Civil', fontweight='bold')

# Gráfica 2: Educación
conteo_education = df['education'].value_counts()
axes[0, 1].bar(conteo_education.index, conteo_education.values)
axes[0, 1].set_title('Distribución por Educación', fontweight='bold')
axes[0, 1].tick_params(axis='x', rotation=45)
axes[0, 1].set_ylabel('Cantidad de registros')

# Gráfica 3: Trabajo
conteo_job = df['job'].value_counts()
axes[1, 0].barh(conteo_job.index, conteo_job.values)
axes[1, 0].set_title('Distribución por Tipo de Trabajo', fontweight='bold')
axes[1, 0].set_xlabel('Cantidad de registros')

# Gráfica 4: Distribución de 'y' por estado civil
if 'y' in df.columns:
    cross_y_marital = pd.crosstab(df['marital'], df['y'])
    cross_y_marital.plot(kind='bar', ax=axes[1, 1])
    axes[1, 1].set_title('Distribución de Suscripción por Estado Civil', fontweight='bold')
    axes[1, 1].set_ylabel('Cantidad de registros')
    axes[1, 1].tick_params(axis='x', rotation=45)
    axes[1, 1].legend(['No', 'Sí'])

plt.tight_layout()
plt.show()

# Análisis resumen
print(f"\n{'='*60}")
print("RESUMEN EJECUTIVO")
print(f"{'='*60}")

print(f"\n1. ESTADO CIVIL:")
print(f"   - Categoría predominante: {reporte_marital.iloc[0]['Categoría']} ({reporte_marital.iloc[0]['Porcentaje']}%)")
print(f"   - Distribución: {', '.join([f'{row.Categoría}: {row.Porcentaje}%' for _, row in reporte_marital.iterrows()])}")

print(f"\n2. EDUCACIÓN:")
print(f"   - Nivel educativo más común: {reporte_education.iloc[0]['Categoría']} ({reporte_education.iloc[0]['Porcentaje']}%)")
print(f"   - Total de categorías educativas: {df['education'].nunique()}")

print(f"\n3. TRABAJO:")
print(f"   - Ocupación más frecuente: {reporte_job.iloc[0]['Categoría']} ({reporte_job.iloc[0]['Porcentaje']}%)")
print(f"   - Distribución laboral variada con {df['job'].nunique()} categorías diferentes")
```

# 7.- Análisis univariado y bivariado basado en la predicción (y/n)

```python
print("=" * 50)
print("ANÁLISIS UNIVARIADO Y BIVARIADO")
print("=" * 50)

# 1. Análisis univariado de la variable objetivo
print("\n1. DISTRIBUCIÓN DE LA VARIABLE OBJETIVO (y):")
distribucion_y = df['y'].value_counts()
porcentaje_y = df['y'].value_counts(normalize=True) * 100

print(f"Total 'no': {distribucion_y['no']} ({porcentaje_y['no']:.1f}%)")
print(f"Total 'yes': {distribucion_y['yes']} ({porcentaje_y['yes']:.1f}%)")

# 2. Análisis bivariado con variables categóricas principales
variables_categoricas = ['job', 'marital', 'education', 'default', 'housing', 'loan']

print("\n2. TASA DE SUSCRIPCIÓN POR CATEGORÍA:")
for variable in variables_categoricas:
    tasa_suscripcion = df.groupby(variable)['y'].apply(lambda x: (x == 'yes').mean() * 100)
    print(f"\n{variable.upper()}:")
    print(tasa_suscripcion.round(1).sort_values(ascending=False))

# 3. Análisis bivariado con variables numéricas
variables_numericas = ['age', 'duration', 'campaign', 'pdays', 'previous', 'emp.var.rate']

print("\n3. ANÁLISIS DE VARIABLES NUMÉRICAS:")
print("Medias por grupo (y='yes' vs y='no'):")
for variable in variables_numericas:
    media_yes = df[df['y'] == 'yes'][variable].mean()
    media_no = df[df['y'] == 'no'][variable].mean()
    print(f"{variable}: yes={media_yes:.1f}, no={media_no:.1f}")

# 4. Visualizaciones rápidas
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Gráfica 1: Distribución de la variable objetivo
df['y'].value_counts().plot(kind='bar', ax=axes[0,0], color=['skyblue', 'lightcoral'])
axes[0,0].set_title('Distribución de Suscripciones (y)')
axes[0,0].set_ylabel('Cantidad')

# Gráfica 2: Tasa de suscripción por trabajo
tasa_job = df.groupby('job')['y'].apply(lambda x: (x == 'yes').mean() * 100)
tasa_job.sort_values().plot(kind='barh', ax=axes[0,1], color='lightgreen')
axes[0,1].set_title('Tasa de Suscripción por Trabajo')
axes[0,1].set_xlabel('Tasa de Sí (%)')

# Gráfica 3: Edad vs Suscripción
df.boxplot(column='age', by='y', ax=axes[1,0])
axes[1,0].set_title('Distribución de Edad por Suscripción')
axes[1,0].set_ylabel('Edad')

# Gráfica 4: Duración vs Suscripción
df.boxplot(column='duration', by='y', ax=axes[1,1])
axes[1,1].set_title('Duración de Llamada por Suscripción')
axes[1,1].set_ylabel('Duración (segundos)')

plt.tight_layout()
plt.show()

# 5. Análisis de correlación con variable objetivo (para variables numéricas)
print("\n4. CORRELACIÓN CON VARIABLE OBJETIVO:")
# Convertir y a numérico para correlación
df_numeric = df.copy()
df_numeric['y_numeric'] = df_numeric['y'].map({'no': 0, 'yes': 1})

correlaciones = df_numeric[variables_numericas + ['y_numeric']].corr()['y_numeric'].sort_values(ascending=False)
print(correlaciones.round(3))

print("\n5. INSIGHTS PRINCIPALES:")
print(f"- Clase desbalanceada: sólo {porcentaje_y['yes']:.1f}% de suscripciones exitosas")
print(f"- Variable más correlacionada: {correlaciones.index[1]} ({correlaciones.iloc[1]:.3f})")
print(f"- Duración promedio mayor en suscripciones exitosas")
```

# 8.- Dividir dataset entre training y test (70:30)

```python
print("=" * 50)
print("DIVISIÓN DEL DATASET - TRAINING Y TEST")
print("=" * 50)

# Importar librería necesaria
from sklearn.model_selection import train_test_split

# Verificar que tenemos la variable objetivo
print("Variable objetivo 'y' disponible:", 'y' in df.columns)

# Separar variables predictoras (X) y variable objetivo (y)
X = df.drop('y', axis=1)  # Todas las columnas excepto la objetivo
y = df['y']  # Variable objetivo

print(f"\nDimensiones de X: {X.shape}")
print(f"Dimensiones de y: {y.shape}")

# Dividir en training y test (70% training, 30% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.3,          # 30% para test
    random_state=42,        # Semilla para reproducibilidad
    stratify=y              # Mantener proporción de clases en ambos sets
)

print(f"\n✅ División completada:")
print(f"   - X_train: {X_train.shape} ({len(X_train)/len(X)*100:.1f}%)")
print(f"   - X_test:  {X_test.shape} ({len(X_test)/len(X)*100:.1f}%)")
print(f"   - y_train: {y_train.shape}")
print(f"   - y_test:  {y_test.shape}")

# Verificar distribución de clases en ambos sets
print(f"\n📊 Distribución de clases:")
print("Training set:")
print(y_train.value_counts(normalize=True).round(3))
print("\nTest set:")
print(y_test.value_counts(normalize=True).round(3))

print(f"\n🎯 Proporción mantenida: {y_train.value_counts(normalize=True)['yes']*100:.1f}% de 'yes' en ambos sets")
```
