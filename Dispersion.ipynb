```python
import pandas as pd
import numpy as np
import os
import plotly.express as px
```

```python
df = pd.read_csv('C:/Users/Rub√©n/Documents/DATA ANALIST/netflix_titles.csv')
df.sample(5)
```

```python
df = df.drop(columns=['Unnamed: 0'])
df['duration_num'] = df['duration_num'].round().astype(int)
```

```python
df_movies_year = df.groupby('release_year').size().rename('movies').reset_index()
df_movies_year
```

```python
fig = px.bar(df_movies_year, x = 'release_year', y= 'movies', title= 'Series y Peliculas por a√±o')
fig.show()
```

```python
df['num_words_title'] = df ['title'].str.split().str.len()
df[['num_words_title']].sample(5)
```

```python
df[['release_year','num_words_title']].sample(5)
```

```python
df_corr = df[['release_year', 'duration_num', 'num_words_title']].corr()
```

```python
fig = px.imshow(df_corr,text_auto=True)
fig.show()
```

```python
# Facet charts
fig = px.strip(df[df['release_year'] > 2015], x="duration_num", y="num_words_title", color="type", facet_col="release_year")
fig.show()
# Cuatro dimensiones en una
```
