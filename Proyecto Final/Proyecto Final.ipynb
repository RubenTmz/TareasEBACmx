# ---------------------------游MACHINE LEARNING游------------------------------

```python
# Filtrar a침os de entrenamiento
df_ts = df_seguros[df_seguros['A침o'].between(2014, 2025)]

# Sumar delitos por a침o
df_ts['Total_anual'] = df_ts[['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio',
                               'Agosto','Septiembre','Octubre','Noviembre','Diciembre']].sum(axis=1)

# Agrupar por a침o
serie = df_ts.groupby('A침o')['Total_anual'].sum()
serie
```

```python
serie.index = pd.to_datetime(serie.index, format='%Y')
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(serie)
plt.title('Delitos en Nuevo Le칩n (20142025)')
plt.xlabel('A침o')
plt.ylabel('N칰mero de delitos')
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

modelo = ARIMA(serie, order=(1,1,1))
resultado = modelo.fit()
resultado.summary()
```

```python
pred_2026 = resultado.forecast(steps=1)
pred_2026
```

```python
plt.figure()
plt.plot(serie, label='Hist칩rico')
plt.plot(pred_2026.index, pred_2026, marker='o', label='Predicci칩n 2026')
# plt.legend()
plt.title('Predicci칩n de delitos 2026')
plt.show()
```

Se utiliz칩 un modelo ARIMA para analizar la evoluci칩n de los delitos entre 2015 y 2025 en Nuevo Le칩n.
El modelo permiti칩 identificar la tendencia hist칩rica y generar una predicci칩n para 2022, la cual sugiere que el nivel de delitos continuar치 acorde al comportamiento observado en a침os anteriores.

# -----------------------------------------Clustering-------------------------------------------

```python
# Filtrar a침o 2021
df_2021 = df[df['A침o'] == 2021]
```

```python
# Calcular total de delitos por estado
meses = ['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio',
         'Agosto','Septiembre','Octubre','Noviembre','Diciembre']

df_2021['Total_delitos'] = df_2021[meses].sum(axis=1)

df_estado = df_2021.groupby('Entidad')['Total_delitos'].sum().reset_index()
df_estado.head()
```

```python
# Normalizaci칩n de datos
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_estado[['Total_delitos']])
```

```python
# Elegir n칰mero de clusters (M칠todo del codo)
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inercia = []

for k in range(1, 7):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inercia.append(kmeans.inertia_)

plt.figure()
plt.plot(range(1,7), inercia, marker='o')
plt.xlabel('N칰mero de clusters')
plt.ylabel('Inercia')
plt.title('M칠todo del codo')
plt.show()
```

```python
# Aplicar K-means
kmeans = KMeans(n_clusters=3, random_state=42)
df_estado['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# Interpretaci칩n de clusters
df_estado.sort_values('Total_delitos', ascending=False)
```

### Cluster 2  Alta peligrosidad

### Cluster 1  Media

### Cluster 0  Baja

Se aplic칩 el algoritmo K-means para clasificar los estados de M칠xico seg칰n su nivel de peligrosidad en 2021.
Los datos fueron previamente normalizados para evitar sesgos por escala.
El m칠todo del codo indic칩 que tres clusters representan adecuadamente los niveles de incidencia delictiva, permitiendo identificar estados con baja, media y alta peligrosidad.
