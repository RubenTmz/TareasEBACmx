```python
import pycountry
import pytz

# Función para obtener el código ISO del país (para que pytz pueda leerlo)
def get_country_code(name):
    try:
        return pycountry.countries.lookup(name).alpha_2
    except LookupError:
        return None

# Obtener código de país
df["Country_Code"] = df["Nationality"].apply(get_country_code)

# Zona Horaria
df["Zona Horaria"] = df["Country_Code"].apply(lambda code: pytz.country_timezones.get(code, [None])[0] if code else None)

# Agregar columna de capital
df["Capital"] = df["Nationality"].apply(lambda x: capitales.get(x, None))

# Mostrar una muestra
df[["Name", "Nationality", "Country_Code", "Zona Horaria", "Capital"]].head(10)
```

```python
# Generar una lambda que en base al potencial, añada una columna “Candidato” o “Next Year”, para el premio para el balón de oro.  Jugadores con más de 85, son candidatos.
df["Balon_Oro"] = df["Potential"].apply(lambda x: "Candidato" if x > 85 else "Next Year")
df[["Name", "Potential", "Balon_Oro"]].head(10)
```
