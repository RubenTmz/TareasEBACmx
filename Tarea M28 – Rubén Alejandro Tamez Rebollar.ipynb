## 3. Archivos JSON (GOT Subtitles.json)

```python
archivos = [f"C:/Users/Rubén/Documents/season{i}.json" for i in range(1, 8)]

# Lista para guardar los datos
datos = []

# Leer todos los archivos
for archivo in archivos:
    with open(archivo, "r", encoding="utf-8") as f:
        data = json.load(f)
        for capitulo, texto in data.items():
            datos.append({
                "capitulo": capitulo,
                "dialogos": texto
            })

# Convertir a DataFrame
df = pd.DataFrame(datos)

# Mostrar las primeras 10 filas
print(df.head(10))

# Mostrar las últimas 10 filas
print(df.tail(10))

# Guardar archivo nuevo con solo las columnas necesarias
df.to_json("got_filtrado.json", orient="records", force_ascii=False)
```
